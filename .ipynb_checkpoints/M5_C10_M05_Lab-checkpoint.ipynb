### RNN with Attention (Possible solution)

```python
# -*- coding: utf-8 -*-
"""
Wind Turbine Power Prediction with RNNs + Attention

This lab uses an Attention-enhanced Recurrent Neural Network to predict wind turbine power output.
"""

# Step 0: Load libraries & Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, Layer
from tensorflow.keras.callbacks import EarlyStopping
import math

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Step 1: Prepare the Data
df = pd.read_csv('t1.csv')

# Parse datetime and set as index
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%d %m %Y %H:%M')
df = df.sort_values('Date/Time').set_index('Date/Time')

# Check for missing values
print("Missing values:\n", df.isnull().sum())
print("\nShape:", df.shape)
print(df.head())

# Select features and target
feature_cols = ['Wind Speed (m/s)', 'Theoretical_Power_Curve (KWh)', 'Wind Direction (°)']
target_col = 'LV ActivePower (kW)'

# Scale features and target separately
# We scale target separately so we can inverse_transform predictions later
feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

scaled_features = feature_scaler.fit_transform(df[feature_cols])
scaled_target = target_scaler.fit_transform(df[[target_col]])

print("\nScaling complete. Feature shape:", scaled_features.shape)
```

The reason we use two separate scalers is subtle but important — when we eventually want to evaluate our predictions in real kW values, we need to reverse the scaling on the target only. If everything was in one scaler, we couldn't do that cleanly.

Step 2: Creating Sequences

Imagine sliding a frame across your time series:
* Window 1:  rows 0–23  → predict row 24
* Window 2:  rows 1–24  → predict row 25
* Window 3:  rows 2–25  → predict row 26
...and so on

Each window becomes one training sample.
With 50,530 rows and a sequence length of 24, we end up with ~50,506 samples — one for nearly every timestep.

```python
# Step 2: Create sequences
def create_sequences(features, target, seq_length=24):
    """
    Sliding window over the time series.
    X: windows of `seq_length` feature rows
    y: the target value at the next timestep after each window
    """
    X, y = [], []
    for i in range(len(features) - seq_length):
        X.append(features[i : i + seq_length])      # window of past readings
        y.append(target[i + seq_length])             # next power value to predict
    return np.array(X), np.array(y)

SEQ_LENGTH = 24  # 24 timesteps × 10 minutes = 4 hours of history

X, y = create_sequences(scaled_features, scaled_target, SEQ_LENGTH)

print("X shape:", X.shape)   # (samples, timesteps, features)
print("y shape:", y.shape)   # (samples, 1)
```

The output shape (samples, 24, 3) is important — this is the exact format an LSTM expects: batch of sequences, each with 24 timesteps, each timestep having 3 features.


Step 3: Split the Data

For time series, we can't use a random train/test split.

Randomly shuffling would let the model "see the future" during training, which gives falsely optimistic results.

We must split chronologically — train on early 2018, test on later 2018.

```python
# Step 3: Split the data (chronological, not random!)

split = int(0.8 * len(X))   # 80% train, 20% test

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"Training samples:  {len(X_train):,}")
print(f"Test samples:      {len(X_test):,}")
print(f"\nTraining covers up to index {split} (~first 8 months)")
print(f"Test covers the remaining ~2 months")
```

# The "why" before the code

A standard LSTM processes your 24 timesteps and then just hands off the last hidden state to the next layer. That final state has to somehow compress everything it learned from all 24 steps into a single vector. That's a lot to ask of it.

Attention solves this by keeping all 24 LSTM output states and then learning a weighted average of them. 

Think of it like a student reviewing 24 hours of notes before an exam — rather than only reading the last page, attention lets the model decide which hours were most informative and focus there.

The mechanism in three steps:

1. Score each of the 24 timestep outputs with a small learned network → 24 raw scores
2. Softmax those scores → 24 weights that sum to 1 (a probability distribution over timesteps)
3. Weighted sum of the 24 LSTM outputs using those weights → one context vector

```python
# Step 4: Define a Custom Attention Layer
class AttentionLayer(Layer):
    """
    Bahdanau-style additive attention.
    Takes all LSTM hidden states and produces a single context vector
    by learning which timesteps matter most.
    """
    def __init__(self, units=64, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)
        # A small dense network that scores each timestep
        self.W = Dense(units, activation='tanh')   # projects each hidden state
        self.v = Dense(1)                          # collapses to a single score

    def call(self, hidden_states):
        # hidden_states shape: (batch, timesteps, lstm_units)

        # Step 1: Score each timestep
        score = self.v(self.W(hidden_states))
        # score shape: (batch, timesteps, 1)

        # Step 2: Convert scores to weights via softmax
        attention_weights = tf.nn.softmax(score, axis=1)
        # weights shape: (batch, timesteps, 1) — sums to 1 across timesteps

        # Step 3: Weighted sum of LSTM outputs
        context_vector = attention_weights * hidden_states
        context_vector = tf.reduce_sum(context_vector, axis=1)
        # context_vector shape: (batch, lstm_units)

        return context_vector, attention_weights
```

A couple of things worth noting here.

This is a subclassed Keras layer — the same pattern you'd use to build any custom component in TensorFlow. 

The __init__ defines the learnable sub-layers, and call defines what happens when data flows through it. Also notice we return the attention_weights alongside the context vector — that's actually useful later if you want to visualise which timesteps the model paid attention to, which is a nice interpretability bonus.

The W + v combination is the classic Bahdanau additive attention pattern — W projects each hidden state into a shared space, tanh squashes it, then v collapses it to a scalar score. It's deliberately simple but very effective

```python
# Input (batch, 24, 3)
#        ↓
# LSTM — returns ALL 24 hidden states, not just the last one
#        ↓
# AttentionLayer — scores each of the 24 states, returns 1 context vector
#        ↓
# Dropout — regularisation to prevent overfitting
#        ↓
# Dense(64, relu) — learns from the context vector
#        ↓
# Dropout
#        ↓
# Dense(1) — outputs a single scaled power prediction
```

The critical detail is return_sequences=True on the LSTM. By default, an LSTM only returns its final hidden state. We need all 24 hidden states so our attention layer has something to score. That one argument is what makes attention possible.


```python
# Step 5: Build the RNN Model with Attention

def build_attention_model(seq_length, n_features, lstm_units=64, attention_units=64):
    # Define input shape
    inputs = Input(shape=(seq_length, n_features))

    # LSTM layer — return_sequences=True gives us all 24 hidden states
    lstm_out = LSTM(lstm_units, return_sequences=True)(inputs)

    # Attention layer — scores and weighs those 24 states
    context_vector, attention_weights = AttentionLayer(units=attention_units)(lstm_out)

    # Fully connected layers to make the final prediction
    x = Dropout(0.2)(context_vector)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.2)(x)
    output = Dense(1)(x)  # no activation — this is a regression output

    model = Model(inputs=inputs, outputs=output)
    return model

model = build_attention_model(
    seq_length=SEQ_LENGTH,
    n_features=3,
    lstm_units=64,
    attention_units=64
)

model.summary()

```

```python
# Step 6: Compile and Train the Model
# Step 6: Compile and Train the Model

model.compile(
    optimizer='adam',
    loss='mse'        # Mean Squared Error — standard for regression
)

# Early stopping watches validation loss and stops if it stops improving
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,           # stop after 5 epochs with no improvement
    restore_best_weights=True  # roll back to the best epoch automatically
)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=64,
    validation_split=0.1,   # use last 10% of training data as validation
    callbacks=[early_stop],
    verbose=1
)

# Plot training history
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss During Training')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss (scaled)')
plt.legend()
plt.tight_layout()
plt.show()
```

A few design decisions explained:

Adam optimiser — adapts the learning rate automatically per parameter. It's the safe default for most neural networks and works well with LSTMs.

MSE loss — since we're predicting a continuous value, we want to penalise large errors heavily. MSE does that by squaring the error. Note this is on scaled values (0–1 range), so the numbers will look tiny — that's normal.

validation_split=0.1 — importantly, Keras takes this from the end of your training data, which preserves chronological order. It won't randomly sample from the middle.

restore_best_weights=True — without this, early stopping gives you the weights from the last epoch, not the best one. Always include this.

```python
# Evaluation

# Generate predictions on test set
y_pred_scaled = model.predict(X_test)

# Inverse transform back to real kW values
y_pred = target_scaler.inverse_transform(y_pred_scaled)
y_actual = target_scaler.inverse_transform(y_test)

# Calculate metrics
rmse = math.sqrt(mean_squared_error(y_actual, y_pred))
r2 = r2_score(y_actual, y_pred)

print(f"Test RMSE: {rmse:.2f} kW")
print(f"Test R²:   {r2:.4f}")

# Plot predictions vs actual (first 500 points for clarity)
plt.figure(figsize=(12, 5))
plt.plot(y_actual[:500], label='Actual Power', alpha=0.8)
plt.plot(y_pred[:500], label='Predicted Power', alpha=0.8)
plt.title('Actual vs Predicted Wind Turbine Power Output')
plt.xlabel('Timestep (10-min intervals)')
plt.ylabel('Power (kW)')
plt.legend()
plt.tight_layout()
plt.show()

# Scatter plot — perfect model would be a diagonal line
plt.figure(figsize=(6, 6))
plt.scatter(y_actual, y_pred, alpha=0.1, s=1)
plt.plot([y_actual.min(), y_actual.max()],
         [y_actual.min(), y_actual.max()], 'r--', label='Perfect prediction')
plt.xlabel('Actual Power (kW)')
plt.ylabel('Predicted Power (kW)')
plt.title('Predicted vs Actual (Scatter)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Deeper diagnostics

# 1. Error distribution
errors = y_actual.flatten() - y_pred.flatten()

plt.figure(figsize=(8, 6))
plt.hist(errors, bins=100, edgecolor='k', alpha=0.7)
plt.axvline(0, color='red', linestyle='--')
plt.title('Prediction Error Distribution')
plt.xlabel('Error (kW)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

print(f"Mean error:   {errors.mean():.2f} kW  (bias — positive means underpredicting)")
print(f"Std of error: {errors.std():.2f} kW")
print(f"Max overpredict:  {errors.min():.2f} kW")
print(f"Max underpredict: {errors.max():.2f} kW")

# 2. Error vs actual power — reveals where the model struggles
plt.figure(figsize=(8, 6))
plt.scatter(y_actual.flatten(), errors, alpha=0.1, s=1)
plt.axhline(0, color='red', linestyle='--')
plt.title('Error vs Actual Power Output')
plt.xlabel('Actual Power (kW)')
plt.ylabel('Error (kW)')
plt.tight_layout()
plt.show()

# 3. How often are we within 10% of max turbine capacity?
max_power = y_actual.max()
within_10pct = np.mean(np.abs(errors) < 0.10 * max_power) * 100
within_5pct  = np.mean(np.abs(errors) < 0.05 * max_power) * 100
print(f"\nMax observed power: {max_power:.0f} kW")
print(f"Predictions within 10% of max capacity: {within_10pct:.1f}%")
print(f"Predictions within 5% of max capacity:  {within_5pct:.1f}%")
```

```python
# What were actual conditions during the worst predictions?
error_series = pd.Series(errors, name='error')

# Find the worst 20 predictions
worst_idx = np.argsort(np.abs(errors))[-20:]

print("Worst 20 predictions:")
print(f"{'Actual (kW)':>12} {'Predicted (kW)':>15} {'Error (kW)':>12} {'Wind Speed':>12}")
print("-" * 55)
for i in worst_idx:
    # Get original wind speed for this test sample
    # Each test sample ends at index split + SEQ_LENGTH + i
    actual_ws_scaled = X_test[i, -1, 0]  # last timestep, wind speed feature
    actual_ws = feature_scaler.inverse_transform(
        [[actual_ws_scaled, 0, 0]])[0][0]
    print(f"{y_actual[i][0]:>12.1f} {y_pred[i][0]:>15.1f} {errors[i]:>12.1f} {actual_ws:>12.2f} m/s")

# Distribution of actual power in test set
plt.figure(figsize=(10, 4))
plt.hist(y_actual.flatten(), bins=100, edgecolor='k', alpha=0.7)
plt.title('Distribution of Actual Power Values in Test Set')
plt.xlabel('Power (kW)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

The worst errors cluster around either very low actual power (turbine idle but model predicts high) or very high actual power (turbine running hard but something caused a sudden shutdown). The histogram show a spike at zero — turbines sitting idle are notoriously hard to handle without explicit treatment.

This is a data structure problem, not a model failure.

## Reading the scatter plot (Image 1):
The vertical spike of errors at actual power ≈ 0 is the dead giveaway. When the turbine is actually producing near-zero power, the model is predicting hundreds or thousands of kW — those are the -3000 kW errors. The model learned "when wind speed looks like this, power should be high" but couldn't anticipate that the turbine was shut down or idling. There's also a smaller spike at the far right around 3500 kW — that's the rated capacity ceiling where the turbine caps out but the model slightly overshoots.

The broad fan shape across the middle (0–3000 kW actual) is normal regression scatter and actually looks reasonable.

## Reading the error histogram (Image 2):
The bulk of predictions are tightly centred around zero error — that tall peak is the model working well for normal operating conditions. The long left tail stretching to -3500 kW is entirely those zero-power shutdown events being overpredicted.
The core issue: your turbine has ~3600 kW rated capacity but clearly spends time at zero (shutdowns, maintenance, low wind). The model never learned to predict "turbine is off" as a distinct state.

The honest interpretation is that 0.88 R² with this data is actually good — if you filtered out the zero-power anomaly periods, you'd likely see R² jump to 0.93+. You can demonstrate this:

```python
normal_ops = y_actual.flatten() > 50

rmse_normal = math.sqrt(mean_squared_error(
    y_actual.flatten()[normal_ops], 
    y_pred.flatten()[normal_ops]
))
r2_normal = r2_score(
    y_actual.flatten()[normal_ops], 
    y_pred.flatten()[normal_ops]
)

print(f"Full test set:      RMSE={rmse:.1f} kW,  R²={r2:.4f}")
print(f"Normal operations:  RMSE={rmse_normal:.1f} kW,  R²={r2_normal:.4f}")
print(f"\nShutdown periods in test set: {(~normal_ops).sum():,} samples "
      f"({(~normal_ops).mean()*100:.1f}%)")
```

```python
# Check all three operating regimes separately
shutdown = y_actual.flatten() <= 50
rated = y_actual.flatten() >= 3400
normal = ~shutdown & ~rated

print(f"Shutdown  (≤50 kW):   {shutdown.sum():,} samples ({shutdown.mean()*100:.1f}%)")
print(f"Normal operations:    {normal.sum():,} samples ({normal.mean()*100:.1f}%)")
print(f"Rated capacity (≥3400 kW): {rated.sum():,} samples ({rated.mean()*100:.1f}%)")

print(f"\nNormal ops only — RMSE: {math.sqrt(mean_squared_error(y_actual.flatten()[normal], y_pred.flatten()[normal])):.1f} kW, "
      f"R²: {r2_score(y_actual.flatten()[normal], y_pred.flatten()[normal]):.4f}")

print(f"\nMean absolute error by regime:")
print(f"  Shutdown:  {np.mean(np.abs(errors[shutdown])):.1f} kW avg error")
print(f"  Normal:    {np.mean(np.abs(errors[normal])):.1f} kW avg error")  
print(f"  Rated:     {np.mean(np.abs(errors[rated])):.1f} kW avg error")
```

This is a really honest and important result.

The normal operating regime R² of 0.77 is actually lower than the full dataset R² of 0.88 — which is counterintuitive and worth understanding properly.

## What's actually happening:
R² measures how much better the model is than just predicting the mean. The full dataset has enormous variance (power swings from 0 to 3500 kW), so even imperfect predictions look good relative to that baseline. When you isolate normal operations, the range narrows and the bar gets harder to clear — a 459 kW RMSE looks worse against a tighter distribution.

The mean absolute errors tell the real story cleanly.

The model is struggling everywhere roughly equally — shutdown (216 kW), normal (309 kW), rated (304 kW). There's no single villain regime. The error is distributed across all operating conditions.

The root cause is almost certainly the features available. You have wind speed, theoretical power curve, and wind direction — but wind power is affected by air density, turbulence intensity, wake effects from nearby turbines, blade pitch angle, and rotor RPM, none of which are in your dataset. The model is doing its best with incomplete information.

**What worked:**
The attention-enhanced LSTM successfully learned the general wind-power relationship, achieving R²=0.88 on a noisy real-world dataset with no missing values handling or feature engineering required.

**What limited performance:**
Three distinct operating regimes (shutdown, normal, rated capacity) each present differently in the data. 22.7% of test samples were shutdown periods where power is decoupled from wind speed entirely. Feature set doesn't include turbine operational state variables.
